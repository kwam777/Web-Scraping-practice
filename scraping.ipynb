Scrape Premier League Stats 

```python
import requests
```

Nach Requests Import wird mithilfe Requests URL-Daten gescrapet

```python
standings_url = "https://fbref.com/de/wettbewerbe/9/Premier-League-Statistiken"
```

```python
data = requests.get(standings_url)
```

```python
data.text
```

Import Beautiful Soup um Daten zu parsen

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(data.text)
```

```python
standings_table = soup.select('table.stats_table')[0]
```

```python
standings_table
```

Filtern nach allen Elementen mit 'a'

```python
links = standings_table.find_all('a')
```

Nun lassen wir uns nur den "href"-Teil ausgeben weil nur dieser fÃ¼r uns relevant ist. 

```python
links = [l.get("href") for l in links]
```

Ausgabe der Teams

```python
links = [l for l in links if '/mannschaften/' in l]
```

Daten geben nur Mannschaftsnamen aus, wir wollen aber einen Link erstellen so wie es im HTML vorhanden ist fÃ¼r den direkten zugriff Also fÃ¼gen wir den notwendigen link an den Anfang hinzu.

```python
team_urls = [f"https://fbref.com{l}" for l in links]
```

```python
team_urls
```
